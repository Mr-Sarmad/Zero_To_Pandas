```python
import pandas as pd
netflix_raw_df = pd.read_csv('netflix-movies-and-tv-shows/netflix_titles.csv', encoding='latin1')
netflix_raw_df
```

```python
netflix_raw_df.columns
```

```python
selected_columns=['show_id', 'type', 'title', 'director', 'cast', 'country', 'date_added',
       'release_year', 'rating', 'duration', 'listed_in', 'description']
```

```python
len(selected_columns)
```

```python
netflix_df=netflix_raw_df[selected_columns].copy()
netflix_df
```

```python
netflix_df.info()
```

```python
netflix_df.drop(netflix_df[pd.isna(netflix_df['country'])].index, inplace=True)
netflix_df.drop(netflix_df[pd.isna(netflix_df['director'])].index, inplace=True)
netflix_df.drop(netflix_df[netflix_df.type=='TV Show'].index,inplace=True)
netflix_df.drop(netflix_df[(netflix_df['duration'] < '150 min') | (netflix_df['duration'] > '300 min')].index, inplace=True)
```

```python
import numpy as np
netflix_df.where(~(netflix_df.country.str.contains(';', na=False)), np.nan, inplace=True)
```

```python
netflix_df.sample(10)
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

# Country

```python
netflix_df.country.nunique()
```

```python
top_countries=netflix_df.country.value_counts().head(15)
top_countries
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title('top_countries')
sns.barplot(x=top_countries.index, y=top_countries);
```

# release year

```python
release_year=netflix_df.release_year.value_counts().head(15)
release_year
```

```python
plt.figure(figsize=(12, 6))
plt.title('Distribution of Release Years')
plt.hist(netflix_df['release_year'], bins=np.arange(1920, 2025, 5), color='red')
plt.xlabel('Release Year')
plt.ylabel('Frequency');
```

# Listed_in

```python
listed_in=netflix_df.listed_in.value_counts().head(10)
listed_in
```

```python
plt.figure(figsize=(12, 6))
plt.title('Distribution of Listed_in Categories')
plt.pie(listed_in.value_counts(), labels=listed_in.value_counts().index, autopct='%1.1f%%', startangle=180);
```

# Director

```python
famous_director = netflix_df['director'].value_counts().head(10)
famous_director
top_10_directors=famous_director.index
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=netflix_df, y='director', order=top_10_directors, palette='viridis')
plt.title('Top 10 Most Frequent Directors on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Director');
```

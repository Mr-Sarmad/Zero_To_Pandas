### Character Meaning

*"r"* open for reading(default)
*"w"* open for the writing , turncating the file first
*"x"* create a new andopen it for writing 
*"a"* open for writing, appending at the end of file if it exisit
*"b"* binary mode
*"t"* text mode(default)
*"+"* open a disk file for uploading(reading and writing)
*"u"* universal newline mode(diprecated)

*read a file line by line*

```python
with open('./data/loans3.txt', 'r', encoding='utf-8') as file3:
    file3_lines = file3.readlines()
    for line in file3_lines:
        print(line)
```

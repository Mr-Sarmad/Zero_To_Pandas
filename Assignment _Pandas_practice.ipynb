### Merge countries_df with covid_data_df on the location column.

```python
combined_df = countries_df.merge(covid_data_df)
```

```python
combined_df
```

### Add columns tests_per_million, cases_per_million and deaths_per_million into combined_df.



```python
combined_df['tests_per_million'] = combined_df['total_tests'] * 1e6 / combined_df['population']
```

```python
combined_df['cases_per_million'] = combined_df['total_cases']*1e6 / combined_df['population']
```

```python
combined_df['deaths_per_million'] = combined_df['total_deaths'] * 1e6 / combined_df['population']
```

```python
combined_df
```

### Create a dataframe with 10 countires that have highest number of tests per million people.

```python
combined_df['cases_per_million'] = combined_df['total_cases'] / (combined_df['population'] / 1_000_000)
highest_cases_df = combined_df.nlargest(10, 'cases_per_million')
highest_cases_df[['location', 'cases_per_million']]
```

###  Create a dataframe with 10 countires that have highest number of deaths cases per million people?

```python
combined_df['deaths_per_million'] = combined_df['total_deaths'] / (combined_df['population'] / 1_000_000)
highest_deaths_df = combined_df.nlargest(10, 'deaths_per_million')
highest_deaths_df
```

###  Count number of countries that feature in both the lists of "highest number of tests per million" and "highest number of cases per million".

```python
combined_df['tests_per_million'] = combined_df['total_tests'] / (combined_df['population'] / 1_000_000)
top_10_tests_per_million = combined_df.nlargest(10, 'tests_per_million')
combined_df['cases_per_million'] = combined_df['total_cases'] / (combined_df['population'] / 1_000_000)
highest_cases_df = combined_df.nlargest(10, 'cases_per_million')
num_common_countries = (combined_df['location'].isin(top_10_tests_per_million['location']) & 
                        combined_df['location'].isin(highest_cases_df['location'])).sum()
num_common_countries
```

###  Count number of countries that feature in both the lists "20 countries with lowest GDP per capita" and "20 countries with the lowest number of hospital beds per thousand population". Only consider countries with a population higher than 10 million while creating the list.

```python
countries_filtered = combined_df[combined_df['population'] > 10_000_000]
lowest_gdp_df = countries_filtered.nsmallest(20, 'gdp_per_capita')
lowest_beds_df = countries_filtered.nsmallest(20, 'hospital_beds_per_thousand')
common_countries = set(lowest_gdp_df['location']).intersection(set(lowest_beds_df['location']))
num_common_countries = len(common_countries)
num_common_countries
```
